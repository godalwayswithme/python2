````python
Here are the solutions to the programming problems you provided. You can copy these codes into a Google Document, and then share the link as instructed.

### 1. Program to accept percentage and display grade
```python
percentage = float(input("Enter percentage: "))

if percentage > 90:
    grade = 'A'
elif 80 < percentage <= 90:
    grade = 'B'
elif 60 <= percentage <= 80:
    grade = 'C'
else:
    grade = 'D'

print(f"Your grade is: {grade}")
```

### 2. Program to accept cost price of a bike and display road tax
```python
cost_price = float(input("Enter the cost price of the bike in Rs: "))

if cost_price > 100000:
    tax = cost_price * 0.15
elif 50000 < cost_price <= 100000:
    tax = cost_price * 0.10
else:
    tax = cost_price * 0.05

print(f"The road tax to be paid is: Rs {tax:.2f}")
```

### 3. Program to display monuments of a city
```python
city = input("Enter the name of the city (Delhi, Agra, Jaipur): ").strip().lower()

if city == "delhi":
    monument = "Red Fort"
elif city == "agra":
    monument = "Taj Mahal"
elif city == "jaipur":
    monument = "Jal Mahal"
else:
    monument = "No monument found for this city."

print(f"The monument in {city.capitalize()} is: {monument}")
```

### 4. Program to check how many times a given number can be divided by 3
```python
number = int(input("Enter a number: "))
count = 0

while number > 10:
    number /= 3
    count += 1

print(f"The number can be divided by 3 {count} times before it is less than or equal to 10.")
```

### 5. Why and When to Use while Loop in Python
**While Loop Explanation:**
A `while` loop in Python is used for repeated execution of a block of code as long as a specified condition is true. It is particularly useful when the number of iterations is not known beforehand.

**Example:**
```python
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```
**When to Use:**
- When you want to continue executing until a condition changes.
- When the exact number of iterations is not predetermined (like reading user input).

### 6. Nested while loop to print 3 different patterns
```python
# Pattern 1
print("Pattern 1:")
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1

# Pattern 2
print("Pattern 2:")
i = 5
while i >= 1:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i -= 1

# Pattern 3
print("Pattern 3:")
i = 1
while i <= 5:
    j = 1
    while j <= 5 - i:
        print(" ", end=" ")
        j += 1
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

### 7. Reverse a while loop to display numbers from 10 to 1
```python
number = 10
while number >= 1:
    print(number)
    number -= 1
```

### 8. Another example to reverse a while loop to display numbers from 10 to 1
```python
count = 10
while count > 0:
    print(count)
    count -= 1
```

### Next Steps
1. Copy these codes and explanations into a Google Document.
2. Ensure the document is shareable with view access.
3. Share the link according to your assignment instructions.

If you need help with creating the document or anything else, let me know!
````
